# Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

Step 1: Generate a Synthetic Dataset with Correlated Features

```python
import pandas as pd
import numpy as np

# Setting a seed for reproducibility
np.random.seed(1)

# Generate a synthetic dataset with correlated features
num_samples = 100
feature_1 = np.random.normal(50, 10, num_samples)  # Main feature
feature_2 = feature_1 + np.random.normal(0, 5, num_samples)  # Correlated with feature_1
feature_3 = feature_1 * 0.5 + feature_2 * 0.5 + np.random.normal(0, 5, num_samples)  # Correlated with both feature_1 and feature_2
feature_4 = np.random.normal(30, 10, num_samples)  # Less correlated feature

# Create a DataFrame
df = pd.DataFrame({
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3,
    'Feature_4': feature_4
})

# Display the first few rows of the dataset
print("Synthetic Dataset:")
print(df.head())
```

Step 2: Visualize the Correlation Matrix Using a Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)

# Adding titles
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#  Using the given x and y data, generate a 3D surface plot to visualize the function 


```python
#Code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays to create a DataFrame (optional for structured storage)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Setting up the 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Adding color bar for reference
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

# Labeling the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Surface Plot of z = sin(sqrt(x^2 + y^2))")

# Display the plot
plt.show()
```

#  Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
#Code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Setting the seed for reproducibility
np.random.seed(25)

# Creating the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),  # Population in millions
    'GDP': np.random.randint(500, 2000, 5)  # GDP in billions
}
df = pd.DataFrame(data)

# Creating the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population']*10,  # Scale bubble size by population
            alpha=0.6, c='blue', edgecolors="w", linewidth=0.5)

# Adding labels for each country
for i in range(df.shape[0]):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], 
             fontsize=10, ha='center', va='center', color="black")

# Labeling the axes and title
plt.xlabel("GDP (Billions)")
plt.ylabel("Population (Millions)")
plt.title("Country Population and GDP Bubble Chart")

# Display the plot
plt.show()
```

# .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Display Bokeh plots in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
data = {
    'x': np.random.uniform(0, 100, 100),
    'y': np.random.uniform(0, 100, 100),
    'sizes': np.random.randint(10, 40, 100),  # Sizes of the markers
    'colors': np.random.choice(['blue', 'green', 'red', 'purple', 'orange'], 100)  # Random colors
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Create a Bokeh ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot with Variable Sizes and Colors",
           x_axis_label="X values", y_axis_label="Y values",
           plot_width=700, plot_height=500)

# Add a scatter plot to the figure
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display Bokeh plots in the notebook
output_notebook()

# Generate random data and create a histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Data Distribution", 
           x_axis_label="Value", y_axis_label="Frequency", 
           plot_height=400, plot_width=600)

# Add the histogram data as a quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

#  Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)â€¨

x = np.linspace(0, 1, 10)â€¨

y = np.linspace(0, 1, 10)â€¨

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap

# Display Bokeh plots in the notebook
output_notebook()

# Data generation
data_heatmap = np.random.rand(10, 10)  # 10x10 grid with random values
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the grid and data for Bokeh compatibility
xx = xx.flatten()
yy = yy.flatten()
colors = data_heatmap.flatten()

# Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(x=xx, y=yy, colors=colors))

# Create a Bokeh figure
p = figure(title="Heatmap", 
           x_axis_label='X', y_axis_label='Y', 
           plot_height=400, plot_width=400)

# Create a linear color map
color_map = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

# Add the heatmap using rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, color=color_map, line_color='white')

# Show the plot
show(p)
```
